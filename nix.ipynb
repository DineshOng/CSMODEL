```python
marinduque_hh_url = "hpq_hh_marinduque.csv"
marinduque_mem_url = "hpq_mem_marinduque.csv"
palawan_hh_url = "hpq_hh_palawan.csv"
palawan_mem_url = "hpq_mem_palawan.csv"
```

```python
marinduque_hh_df = pd.read_csv(marinduque_hh_url, dtype='unicode')
marinduque_hh_df.shape
```

```python
marinduque_mem_df = pd.read_csv(marinduque_mem_url, dtype='unicode')
marinduque_mem_df.shape
```

```python
marinduque_df = pd.merge(marinduque_hh_df, marinduque_mem_df, how='inner', on=['main.id'])
marinduque_df
```

```python
marinduque_df["island"] = "Marinduque"
marinduque_df.head()
```

```python
palawan_hh_df = pd.read_csv(palawan_hh_url, dtype='unicode')
palawan_hh_df.shape
```

```python
palawan_mem_df = pd.read_csv(palawan_mem_url, dtype='unicode')
palawan_mem_df.shape
```

```python
palawan_df = pd.merge(palawan_hh_df, palawan_mem_df, how='inner', on=['main.id'])
palawan_df
```

```python
palawan_df["island"] = "Palawan"
palawan_df.head()
```

```python
all_df = palawan_df.append(marinduque_df, ignore_index=True)
all_df
```

```python
loc_df = all_df[["mun", "zone", "brgy", "purok", "island"]].drop_duplicates()
loc_df
```

```python
loc_df = loc_df.drop_duplicates()
loc_df
```

```python
loc_df = loc_df.reset_index(drop=True)
#mar_loc_df.reset_index(drop=True, inplace=True)
loc_df["locID"] = loc_df.index
loc_df
```

```python
all_df = pd.merge(all_df, loc_df, how="left", on=["mun", "zone", "brgy", "purok", "island"])
#marinduque_df = pd.merge(marinduque_hh_df, mar_loc_df, how='left', left_on=["mun", "zone", "brgy", "purok"], right_on=["mun", "zone", "brgy", "purok"])
all_df
```
